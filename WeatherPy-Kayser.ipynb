Analysis:

1. When plotting temperature versus latitude, there is a clear trend of hotter temperatures near the equator and cooler temperatures as the latitudes move away in either direction of the equator. 
2. The plot of latitudes versus cloudiness shows no obvious relation between the two variables. The distance from the equator does not affect the cloudiness of the city. 
3. There is a slight tendancy for higher wind speeds away from the equator, though the trend is not as strong as the temperatures. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key + "&q="

# Create lists to save values
cities_clean = []
lats_clean = []
lngs_clean = []
temp = []
humid = []
cloud = []
wind = []
record_count = 0

# Loop to find current weather situation

print("Beginning data retrieval")
print("------------------------")

for city in cities:

    try:
        weather_data = requests.get(url + city.replace(" ", "%20")).json()
        record_count = record_count + 1
        print("Processing record " + str(record_count) + " | " + city)
        print(url + city.replace(" ", "%20"))
        cities_clean.append(weather_data["name"])
        lats_clean.append(weather_data["coord"]["lat"])
        lngs_clean.append(weather_data["coord"]["lon"])
        temp.append(weather_data["main"]["temp"])
        humid.append(weather_data["main"]["humidity"])
        cloud.append(weather_data["clouds"]["all"])
        wind.append(weather_data["wind"]["speed"])
    except KeyError:
        print(f"Sorry, we don't have data for {city}")    

print("------------------------")
print("Data retrieval is complete")
print("------------------------")

```

```python
#Create dataframe to store all vaules
weather_data_df = pd.DataFrame({"City" : cities_clean,
                          "Latitude" : lats_clean,
                          "Longitude" : lngs_clean,
                          "Temperature" : temp,
                          "Humidity" : humid,
                          "Cloudiness" : cloud,
                          "Wind Speeds" : wind})

weather_data_df.head()
weather_data_df.to_csv("output_data/cities.csv")
```

```python
# PLOT 1
plt.scatter(lats_clean, temp, edgecolors="black")
plt.title("City Latitude versus Temperature (F) (9/26/18)")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.grid()
plt.savefig("output_data/LatitudeVTemp.png")
plt.show
```

```python
# PLOT 2
plt.scatter(lats_clean, humid, edgecolors="black")
plt.title("City Latitude versus Humidity (%) (9/26/18)")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.grid()
plt.savefig("output_data/LatitudeVHumidity.png")
plt.show
```

```python
# PLOT 3
plt.scatter(lats_clean, cloud, edgecolors="black")
plt.title("City Latitude versus Cloudiness (%) (9/26/18)")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.grid()
plt.savefig("output_data/LatitudeVCloud.png")
plt.show
```
